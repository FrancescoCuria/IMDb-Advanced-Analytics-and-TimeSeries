```python
# Supponiamo che `imdb_ds` e `country_code` siano già definiti
# country_code deve avere 'Code' come indice e una colonna 'Region' che mappa i paesi alle regioni

# remove all space in all entries in the column 'Region'
country_code['Region'] = country_code['Region'].str.replace(' ', '')

# Funzione per mappare una lista di paesi alle regioni
def map_countries_to_regions(country_list, country_code):
    regions = set()  # Utilizziamo un set per evitare duplicati
    for country in country_list:
        if country in country_code.index:  # Controlla se il paese è presente in country_code
            regions.add(country_code.loc[country, 'Region'])
    return list(regions) if regions else ["Unknown"]  # Se non ci sono regioni, restituisci "unknown"

# Funzione per convertire una stringa di paesi in una lista
def parse_country_list(country_string):
    if pd.notna(country_string):  # Controlla se il valore non è NaN
        # Rimuovi i caratteri indesiderati e dividi sulla virgola
        return [c.strip() for c in country_string.replace("[", "").replace("]", "").replace("'", "").split(",")]
    return ["Unknown"]  # Se il valore è NaN, restituisci "unknown"

# Prepara le colonne `countryOfOrigin` e `regions` come liste di paesi
imdb_ds['countryOfOrigin'] = imdb_ds['countryOfOrigin'].apply(parse_country_list)
imdb_ds['regions'] = imdb_ds['regions'].apply(parse_country_list)

# Genera la colonna `origin_nomeRegione`
imdb_ds['regionOfOrigin'] = imdb_ds['countryOfOrigin'].apply(
    lambda countries: map_countries_to_regions(countries, country_code)
)

# Genera la colonna `regions_nomeRegione`
imdb_ds['regionAvailability'] = imdb_ds['regions'].apply(
    lambda regions: map_countries_to_regions(regions, country_code)
)
```

```python
def check_subset(ds, col1, col2):
    """
    Check if the elements of col1 are a subset of the elements of col2 for rows where both are not empty.
    Print the number of rows where col1 is a subset of col2 and the number of rows where it is not.
    
    Parameters:
    ds (pd.DataFrame): The dataset.
    col1 (str): The name of the first column.
    col2 (str): The name of the second column.
    """
    subset_count = 0
    not_subset_count = 0

    for index, row in ds.iterrows():
        # Check if both columns are not empty or NaN
        if isinstance(row[col1], list) and isinstance(row[col2], list):
            set1 = set(row[col1])
            set2 = set(row[col2])
            if set1.issubset(set2):
                subset_count += 1
            else:
                not_subset_count += 1

    print(f"Number of rows where {col1} is a subset of {col2}: {subset_count}")
    print(f"Number of rows where {col1} is not a subset of {col2}: {not_subset_count}")

check_subset(imdb_ds, "regionOfOrigin", "regionAvailability")

check_subset(imdb_ds, "regionAvailability", "regionOfOrigin")
```

```python
def check_subset_unknown(ds, col1, col2):
    """
    Check if the elements of col1 are a subset of the elements of col2 for rows where both are not empty.
    Exclude rows where either column contains "unknown" and count those separately.
    Print the number of rows where col1 is a subset of col2, not a subset, equal to col2, and the number of excluded rows.
    
    Parameters:
    ds (pd.DataFrame): The dataset.
    col1 (str): The name of the first column.
    col2 (str): The name of the second column.
    """
    subset_count = 0
    not_subset_count = 0
    equal_count = 0
    unknown_count = 0

    for index, row in ds.iterrows():
        # Check if both columns are lists
        if isinstance(row[col1], list) and isinstance(row[col2], list):
            set1 = set(row[col1])
            set2 = set(row[col2])
            
            # Exclude rows with "unknown" in either column
            if "Unknown" in set1 or "Unknown" in set2:
                unknown_count += 1
                continue
            
            # Check if the sets are equal
            if set1 == set2:
                equal_count += 1
            # Check subset relationship
            elif set1.issubset(set2):
                subset_count += 1
            else:
                not_subset_count += 1

    print(f"Rows where {col1} is equal to {col2}: {equal_count}")
    print(f"Rows where {col1} is smaller (subset) of {col2}: {subset_count}")
    print(f"Rows where {col1} is greater (not a subset) of {col2}: {not_subset_count}")
    print(f"Number of tuples with at least one 'unknown': {unknown_count}")

# Esempio di utilizzo
print("----------")
check_subset_unknown(imdb_ds, "regionOfOrigin", "regionAvailability")
print("----------")
check_subset_unknown(imdb_ds, "regionAvailability", "regionOfOrigin")
print("----------")
```

```python
def calculate_difference(ds, col1, col2):
    """
    Calcola il numero totale di elementi di differenza tra col1 e col2
    per le righe in cui col1 non è un sottoinsieme di col2.

    Parameters:
    ds (pd.DataFrame): Il dataset.
    col1 (str): Il nome della prima colonna.
    col2 (str): Il nome della seconda colonna.

    Returns:
    int: Il numero totale di elementi di differenza.
    """
    total_difference = 0

    for index, row in ds.iterrows():
        # Check if both columns are lists
        if isinstance(row[col1], list) and isinstance(row[col2], list):
            set1 = set(row[col1])
            set2 = set(row[col2])
            
            # Exclude rows with "unknown" in either column
            if "Unknown" in set1 or "Unknown" in set2:
                continue
            
            # Check if col1 is not a subset of col2
            if not set1.issubset(set2):
                # Calculate the difference and add to the total
                difference = set1 - set2
                total_difference += len(difference)

    print(f"Total number of elements in {col1} not in {col2}: {total_difference}")
    return total_difference

# Esempio di utilizzo
print("----------")
calculate_difference(imdb_ds, "regionOfOrigin", "regionAvailability")
print("----------")
    # calculate_difference(imdb_ds, "regionAvailability", "regionOfOrigin")
print("----------")
```

```python
from collections import Counter

def calculate_difference(ds, col1, col2):
    """
    Calcola il numero di righe con un certo numero di elementi di differenza tra col1 e col2
    per le righe in cui col1 non è un sottoinsieme di col2.

    Parameters:
    ds (pd.DataFrame): Il dataset.
    col1 (str): Il nome della prima colonna.
    col2 (str): Il nome della seconda colonna.

    Returns:
    dict: Un dizionario con il numero di differenze come chiave e il conteggio delle righe come valore.
    """
    difference_counts = Counter()

    for index, row in ds.iterrows():
        # Check if both columns are lists
        if isinstance(row[col1], list) and isinstance(row[col2], list):
            set1 = set(row[col1])
            set2 = set(row[col2])
            
            # Exclude rows with "unknown" in either column
            if "Unknown" in set1 or "Unknown" in set2:
                continue
            
            # Check if col1 is not a subset of col2
            if not set1.issubset(set2):
                # Calculate the difference
                difference = set1 - set2
                difference_counts[len(difference)] += 1

    # Print the results
    for diff_count, row_count in sorted(difference_counts.items()):
        print(f"Number of rows with {diff_count} element(s) of difference: {row_count}")

    return difference_counts

# Esempio di utilizzo
print("----------")
print("Check rows where regionOfOrigin is not a subset of regionAvailability and then computes how many elements are different")
calculate_difference(imdb_ds, "regionOfOrigin", "regionAvailability")
print("----------")
```

```python
from collections import Counter

def calculate_difference(ds, col1, col2, output_file="difference_results.txt"):
    """
    Calcola il numero di righe con un certo numero di elementi di differenza tra col1 e col2
    per le righe in cui col1 non è un sottoinsieme di col2. Inoltre, stampa le regioni che differiscono
    e il numero di righe in cui ciascuna regione è presente nella differenza. Salva i risultati in un file di testo.

    Parameters:
    ds (pd.DataFrame): Il dataset.
    col1 (str): Il nome della prima colonna.
    col2 (str): Il nome della seconda colonna.
    output_file (str): Il percorso del file di output per salvare i risultati.

    Returns:
    None
    """
    difference_counts = Counter()
    detailed_differences = {}

    for index, row in ds.iterrows():
        # Check if both columns are lists
        if isinstance(row[col1], list) and isinstance(row[col2], list):
            set1 = set(row[col1])
            set2 = set(row[col2])
            
            # Exclude rows with "unknown" in either column
            if "Unknown" in set1 or "Unknown" in set2:
                continue
            
            # Check if col1 is not a subset of col2
            if not set1.issubset(set2):
                # Calculate the difference
                difference = set1 - set2
                diff_size = len(difference)
                difference_counts[diff_size] += 1
                
                # Save the specific differences
                difference_tuple = tuple(sorted(difference))  # Sort for consistency
                if difference_tuple not in detailed_differences:
                    detailed_differences[difference_tuple] = 0
                detailed_differences[difference_tuple] += 1

    # Write results to a file
    with open(output_file, "w") as f:
        f.write("Number of rows with specific differences:\n")
        for diff_count, row_count in sorted(difference_counts.items()):
            f.write(f"Number of rows with {diff_count} element(s) of difference: {row_count}\n")
        
        f.write("\nDetailed differences and their counts:\n")
        for difference, count in sorted(detailed_differences.items(), key=lambda x: (-len(x[0]), x[1])):
            f.write(f"{difference}: {count}\n")

    print(f"Results saved to {output_file}")
```
